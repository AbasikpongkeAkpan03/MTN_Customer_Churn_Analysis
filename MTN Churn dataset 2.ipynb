```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure


```

```python
# Read in the data
df = pd.read_csv(r"C:\Users\hp\Downloads\Data sets\mtn_customer_churn.csv")
```

```python
df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Confirming data types for each column

df.dtypes
```

```python
df
```

```python
# Creating Column for 'Month'
df['Month'] = df['Date of Purchase'].astype(str).str[:3]
df.head()
```

```python
# Order table by Data Usage
df.sort_values(by=['Data Usage'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop duplicates
df.drop_duplicates()
```

```python
# Exploratory Data Analysis,EDA-Correlation
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
       df_numerized[col_name] = df_numerized[col_name].astype('category')
       df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df_numerized.corr()
```

```python
Correlation_matrix = df_numerized.corr()
sns.heatmap(Correlation_matrix, annot=True)
plt.title('Correlation Matrix or Numeric Features')
plt.xlabel('Data Features')
plt.ylabel('Data Features')

plt.show()
```
